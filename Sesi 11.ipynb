# Underfitting dan Overfitting




Data sebaiknya dipisah menjadi data training dan data testing

1. Metric pada Data Training jauh lebih baik dibandingkan pada data testing (overfitting)
2. Metric pada Data Training jauh lebih buruk dibandingkan pada data testing (underfitting)
3. Metric pada Data Training sama baiknya dibandingkan pada data testing (well-fit)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Linear Regression

```python
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1,1))
y = np.array([5, 20, 14, 32, 22, 38])
```

```python
print(x)
print(y)
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
# R-Squared Semakin Mendekati 1 maka semakin bagus
r_sq = model.score(x,y)
r_sq
```

```python
print('intercept :', model.intercept_)
print('slope :', model.coef_)
```

y = mx + b

y adalah target 

m adalah slope 

b adalah intercept

```python
print('Rumus dari Model adalah y = ',model.coef_[0], 'x +', model.intercept_)
```

```python
y_pred = model.predict(x)
y_pred
```

```python
print(x)
print(y)
```

```python
plt.scatter(x,y, alpha =0.5)
plt.plot(x,y_pred)

plt.title('Scatter Plot x dan y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
y_new_pred = model.predict(x_new)
print(y_new_pred)
```

## Advanced Linear Regression with Statsmodel

```python
import statsmodels.api as sm
```

```python
x = np.array([[0,1],
            [5,1],
            [15,2],
              [25,5],
              [35,11],
              [45,15],
              [55,34],
              [60,35]])
y = np.array([4,5,20,14,32,22,38,43])
```

```python
x = sm.add_constant(x)
```

```python
x
```

```python
model_stat = sm.OLS(y,x)
```

```python
result = model_stat.fit()
```

```python
print(result.summary())
```

formula regresi : y = 5.5226 + 0.447x1 +0.2550x2

```python
print('RSquared =',result.rsquared)
```

```python
result.params
```

```python
res_summary = result.summary()
res_summary
```

```python
res_summary.tables[0]
```

```python
y_pred = result.predict(x)
```

```python
mse = mean_squared_error(y,y_pred)
```

```python
rmse = np.sqrt(mse)
rmse
```

## Data Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
linreg.score(X_train,y_train)
```

```python
y_pred = linreg.predict(X_test)
```

```python
y_pred_train = linreg.predict(X_train)
y_pred_test = linreg.predict(X_test)
```

```python
mse_train = mean_squared_error(y_train, y_pred_train)
mse_test = mean_squared_error(y_test, y_pred_test)
rmse_train = np.sqrt(mse_test)
rmse_test = np.sqrt(mse_train)
print('train mse:', mse_train)
print('test mse:', mse_test)
print('train rmse:', rmse_train)
print('test rmse:', rmse_test)
```

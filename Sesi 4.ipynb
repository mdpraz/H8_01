# Buat function kalkulator dengan operator :
- 'x' sebagai kali
- ':' sebagai bagi
- '+' sebagai tambah
- '-' sebagai kurang
- '^' sebagai pangkat

```python
operasi = input('Pilih operasi (+/-/x/:/^): ')

if operasi == '+':
    print('Anda memilih penjumlahan')
elif operasi == '-':
    print('Anda memilih pengurangan')
elif operasi == 'x':
    print('Anda memilih perkalian')
elif operasi == ':':
    print('Anda memilih pembagian')
elif operasi == '^':
    print('Anda memilih perpangkatan')
else:
    print('Tidak valid')

bilangan_1 = int(input('Masukkan bilangan pertama: '))
bilangan_2 = int(input('Masukkan bilangan kedua: '))


if operasi == '+':
  hasil = bilangan_1 + bilangan_2
  print(f'Hasil operasi dari {bilangan_1} + {bilangan_2} = {hasil}')
elif operasi == '-':
  hasil = bilangan_1 - bilangan_2
  print(f'Hasil operasi dari {bilangan_1} - {bilangan_2} = {hasil}')
elif operasi == 'x':
  hasil = bilangan_1 * bilangan_2
  print(f'Hasil operasi dari {bilangan_1} * {bilangan_2} = {hasil}')
elif operasi == ':':
  hasil = bilangan_1 / bilangan_2
  print(f'Hasil operasi dari {bilangan_1} / {bilangan_2} = {hasil}')
elif operasi == '^':
  hasil = bilangan_1 ** bilangan_2
  print(f'Hasil operasi dari {bilangan_1} ** {bilangan_2} = {hasil}')
else:
  print('Tidak valid')
```

# Membuat Generator Bilangan Prima

```python
lower_value = int(input('Masukkan nilai awal: '))
upper_value = int(input('Masukkan nilai akhir: '))

print('Bilangan Prima di Interval Angka Tersebut Adalah : ')
for a in range(lower_value,upper_value+1):
    if a > 1:
        for i in range (2, a):#Filtering buat bilangan prima
            if (a % i) == 0:
                break
        else:
            print(a)#untuk kita dapat bilangan prima kondisi (a % i)==0 harus dilanggar
```

```python
lowest = int(input('Please, enter the lowest range value : '))
upper = int(input('Please, enter the upper range value : '))

print('The prime numbers in the range are : ')
for i in range(lowest, upper+1):
    if i == 1:
        continue
    if i == 2 or i == 3 or i == 5 or i == 7:
        print(i)
    if i % 2 != 0 :
        if i % 3 != 0 :
            if i % 5 != 0 :
                if i % 7 != 0:
                    print(i)
            
```

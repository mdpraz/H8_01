```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np
```

## Naive Bayes

```python
#Feature 1
weather = ['Sunny','Sunny','Overcast','Rainy','Rainy','Rainy','Overcast','Sunny','Sunny','Rainy',
          'Sunny','Overcast','Overcast','Rainy']
# Feature 2
temp = ['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool',
       'Mild','Mild','Mild','Hot','Mild']

# Label atau target
play = ['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']

print(len(weather),len(temp),len(play))
```

## Encode Features

```python
le = LabelEncoder()

w_enc = le.fit_transform(weather)
t_enc = le.fit_transform(temp)
p_enc = le.fit_transform(play)
```

```python
w_enc
```

```python
t_enc
```

```python
p_enc
```

```python
features = list(zip(w_enc,t_enc))
features
```

```python
np.array(features).shape
```

### Model buiding

```python
from sklearn.naive_bayes import GaussianNB

modelNB = GaussianNB()
```

## Model Fitting

```python
modelNB.fit(features,p_enc)

modelNB.predict([[0,2]]) # Dimana 0 Overcast, 2 Mild
```

```python
modelNB.score(features,p_enc)
```

## Decision Tree

### Decision Tree in SKLEARN

```python
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split
```

```python
col_names = ['pregnant','glucose','bp','skin','insulin','bmi','pedigree','age','label']

pima = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv', header=None, names=col_names)
pima.head()
```

```python
pima.drop(0, inplace=True)
```

```python
pima.info()
```

```python
pima.head()
```

```python
pima.isna().sum()
```

```python
for i in pima.columns:
    pima[i]=pima[i].astype(float)
```

```python
pima.head()
```

```python
feature_cols = ['pregnant','insulin','bmi','age','glucose','bp','pedigree']

X = pima[feature_cols]
y = pima['label']
```

```python
# Split Data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=2024)
```

```python
# Build Model

dtc = DecisionTreeClassifier()

dtc.fit(X_train,y_train)

y_pred_train = dtc.predict(X_train)
y_pred_test = dtc.predict(X_test)
```

```python
print('train set \n',classification_report(y_train,y_pred_train))

print('test set \n',classification_report(y_test,y_pred_test))
```

Model diatas dapat dikatakan overfitting

```python
X = df_iris.drop('species', axis=1)
y = df_iris['species']

# Splitting Data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=2024)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=100)

rfc.fit(X_train,y_train)

y_pred_train = rfc.predict(X_train)
y_pred_test = rfc.predict(X_test)
```

```python
print('Train train \n', classification_report(y_train,y_pred_train))
print('Train set \n', classification_report(y_test,y_pred_test))
```

```python
train.LoanAmount.fillna(train.LoanAmount.mean(), inplace=True)
```

```python
train.Loan_Amount_Term.fillna(train.Loan_Amount_Term.mean(), inplace=True)
train.Credit_History.fillna(train.Credit_History.mean(), inplace=True)
```

```python
train.isna().sum()
```

```python
train.hist(figsize=(10,7))
```

```python
train['Loan_Amount_Term']=np.log(train['Loan_Amount_Term'])
```

```python
train['Loan_Amount_Term'].hist()
```

```python
X = train.drop(['Loan_Status','Loan_ID'],axis=1)
y = train.Loan_Status
```

One-Hot Encoding

```python
X = pd.get_dummies(X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=2024)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
# Decision Tree
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
# Random Forest
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
```

```python
# SVM
svc_model = SVC()
svc_model.fit(X_train,y_train)
```

```python
# Naive Bayes
nb = GaussianNB()
nb.fit(X_train,y_train)
```

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
y_pred_lr = lr.predict(X_test)
y_pred_dt = dtc.predict(X_test)
y_pred_rf = rfc.predict(X_test)
y_pred_sv = svc_model.predict(X_test)
y_pred_knn = knn.predict(X_test)
```

```python
print('Accuracy Score')
print('Logistic Regression', accuracy_score(y_test,y_pred_lr))
print('Decision Tree', accuracy_score(y_test,y_pred_dt))
print('Random Forest', accuracy_score(y_test,y_pred_rf))
print('SVM', accuracy_score(y_test,y_pred_sv))
print('Naive Bayes', accuracy_score(y_test,y_pred_knn))
```
